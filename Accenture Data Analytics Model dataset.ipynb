```python
import pandas as pd
import numpy as np
```

```python
#read the datasets and view the first five rows
reactions = pd.read_csv("Reactions (1).csv")
reactiontypes = pd.read_csv("ReactionTypes (1).csv")
content = pd.read_csv("Content (1).csv")
profile= pd.read_csv('Profile (1).csv')
session = pd.read_csv('Session (1).csv')

reactions.head()
```

```python
#drop the unnamed column
content.drop('Unnamed: 0', axis=1, inplace=True)
reactions.drop('Unnamed: 0', axis=1, inplace=True)
reactiontypes.drop('Unnamed: 0', axis=1, inplace=True)
profile.drop('Unnamed: 0',axis=1,inplace=True)
session.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
#merge the datasets
df = content.merge(profile, how='left', on='User ID')
df = df.merge(session, how='left', on='User ID')
df = df.merge(reactions, how='left', on='Content ID')
df = df.rename(columns={'Type_x':'Cont_Type','Type_y':'Type'})
df = df.merge(reactiontypes, how='left', on='Type')
df = df.rename(columns={'Type':'React_Type'})

df
```

```python
#analyse the df
df.info()
```

```python
#count how many null values are present
df.isnull().sum()
```

```python
#drop the columns with null values
df.drop(['URL','Interests', 'User ID_y'],axis=1,inplace=True)
```

```python
df = df.dropna()
```
